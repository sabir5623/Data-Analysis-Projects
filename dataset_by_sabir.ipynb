```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:/Users/Pc/Downloads/data.csv")
df
```

```python
# dropping irreravant columns.
df1=df.drop(['Engine Fuel Type','Number of Doors',
             'Market Category','Vehicle Size','Vehicle Style','Popularity'],axis=1)
df1
```

```python
# rename columns name.
df2=df1.rename(columns={'Engine HP':'HP','Engine Cylinders':'Cylinders',
                        'Transmission Type':'Transmission','Driven_Wheels':'Drive Mode',
                        'highway MPG':'MPG-H','city mpg':'MPG-C','MSRP':'Price'})
df2
```

```python
# finding duplicate values in rows.
dc_values=df2[df2.duplicated()]
print("total duplicate values in rows are: ",len(dc_values))
dc_values
```

```python
# dropping duplicate rows.
df3=df2.drop_duplicates()
df3
```

```python
# finding missing or null values.
df3.isnull().sum()
```

```python
# dropping missing or null values.
df4=df3.dropna()
df4
```

```python
df4.count()
```

```python
# detecting outliers
```

```python
sns.boxplot(x=df4["Price"]);
```

```python
sns.boxplot(x=df4["HP"]);
```

```python
sns.boxplot(x=df4["Cylinders"]);
```

```python
#finding quantile
df4.quantile()
```

```python
Q1=df4.quantile(0.25)
Q3=df4.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
df5= df4[~((df4 <(Q1-1.5*IQR))|(df4 >(Q3+1.5*IQR))).any(axis=1)]
df5.shape
```

```python
df5
```

```python
# plot different features against one another (scatter), against frequency (histogram) 
```

```python
#histogram
df5.Make.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
plt.title("numbers of car by make")
plt.xlabel("Make")
plt.ylabel("Number of cars")
plt.show()
```

```python
#heat maps
plt.figure(figsize=(10,5))
correlation=df5.corr()
sns.heatmap(correlation,cmap="BrBG",annot=True)
correlation
```

```python
#scatterplot
plt.scatter(df5["HP"],df5["Price"])
plt.xlabel("HP")
plt.ylabel("Price")
plt.show()
```
